```python
import cv2
import psycopg2
from ultralytics import YOLO
from datetime import datetime
```

```python
# Подключаемся. autocommit=True значит, что данные сразу записываются "намертво"
conn = psycopg2.connect("dbname=user43 user=user43 password=m5q3x8tpc7vn host=2.nntc.nnov.ru port=5402")
conn.autocommit = True
cursor = conn.cursor()

model = YOLO('yolo11s.pt')
cap = cv2.VideoCapture('video.mp4')

# Словарь для хранения прошлых позиций (нужен для расчета скорости)
prev_positions = {} 

while cap.isOpened():
    success, frame = cap.read()
    if not success: break

    # Запускаем трекер. persist=True заставляет YOLO помнить ID машины между кадрами
    results = model.track(frame, persist=True, verbose=False)

    if results[0].boxes.id is not None:
        # .cpu().numpy() переносит данные из видеокарты в обычную память для Python
        boxes = results[0].boxes.xyxy.cpu().numpy() 
        ids = results[0].boxes.id.int().cpu().numpy()
        classes = results[0].boxes.cls.int().cpu().numpy()

        for i in range(len(boxes)):
            x1, y1, x2, y2 = boxes[i]
            t_id = int(ids[i])
            cls_name = model.names[int(classes[i])]

            # 1. Сначала работаем со справочником. 
            # ON CONFLICT говорит: "Если такой тип уже есть, просто обнови время последнего визита"
            cursor.execute("""
                INSERT INTO vehicle_types (name_mode, first_seen, last_seen)
                VALUES (%s, %s, %s)
                ON CONFLICT (name_mode) DO UPDATE SET last_seen = EXCLUDED.last_seen
                RETURNING id;
            """, (cls_name, datetime.now(), datetime.now()))
            
            # Запоминаем ID типа, который выдала база
            v_type_id = cursor.fetchone()[0]

            # 2. Считаем скорость (простая версия: разница координат)
            # Если машина была в прошлом кадре, считаем расстояние
            current_speed = 0.0
            cx, cy = (x1 + x2) / 2, (y1 + y2) / 2 # Центр машины
            if t_id in prev_positions:
                # Расстояние в пикселях. В реальности нужно умножать на коэффициент PPM
                dist = ((cx - prev_positions[t_id][0])**2 + (cy - prev_positions[t_id][1])**2)**0.5
                current_speed = dist * 0.5 # Заглушка коэффициента

            prev_positions[t_id] = (cx, cy)

            # 3. Пишем в "сырую" таблицу. Представление (VIEW) само потом очистит эти данные.
            cursor.execute("""
                INSERT INTO traffic_raw (detection_time, vehicle_type_id, speed, track_id, x1, y1, x2, y2)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
            """, (datetime.now(), v_type_id, current_speed, t_id, x1, y1, x2, y2))

cap.release()
conn.close()
```
